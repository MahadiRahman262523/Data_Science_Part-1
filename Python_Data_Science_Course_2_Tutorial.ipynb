# Linear Regression

```python
import pandas as pd
```

```python
iris = pd.read_csv("iris.csv")
```

```python
iris.head()
```

```python
iris['SepalWidthCm']>4
```

```python
iris[iris['SepalWidthCm']>4]
```

```python
iris[iris['PetalWidthCm']>1]
```

```python
iris[iris['PetalWidthCm']>2]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.scatterplot(x="SepalLengthCm",y="PetalLengthCm",data=iris)
plt.show()
```

```python
sns.scatterplot(x="SepalLengthCm",y="PetalLengthCm",data=iris,hue="Species")
plt.show()
```

```python
iris.head()
```

```python
x = iris[["SepalWidthCm"]]
```

```python
y = iris[["SepalLengthCm"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test,y_pred)
```

```python
y = iris[["SepalLengthCm"]]
```

```python
x = iris[["SepalWidthCm","PetalLengthCm","PetalWidthCm"]]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
lr2 = LinearRegression()
```

```python
lr2.fit(x_train,y_train)
```

```python
y_pred2 = lr2.predict(x_test)
```

```python
mean_squared_error(y_test,y_pred2)
```
